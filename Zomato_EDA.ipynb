```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#TOP 3 COUNTRIES USING ZOMATO
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

```python
##OBSERVATION:- Maximum records of zomato is from INDIA. after that United States and United Kingdom.
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
##OBSERVATION:-
#Rating from 4.5 to 4.9 it is Excellent
#Rating from 4.0 to 4.4 it is Very Good
#Rating from 3.5 to 3.9 it is Good
#Rating from 2.5 to 3.4 it is Average
#Rating from 1.8 to 2.4 it is Poor
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, hue='Rating color', palette=['white','red','orange','yellow','green','green'])
```

```python
##OBSERVATION :- Not Rated count are very high and Maximum number of rating are between 2.5 to 3.4
```

```python
sns.countplot(x='Rating color', data=ratings, palette=['white','red','orange','yellow','green','green'])
```
